```python
import tensorflow as tf
from utils import get_data, augment_data
batch_size = 10

print('\nProcessing data...')
train_imgs, train_msks, valid_imgs, valid_msks, _, _ = \
    get_data(path_train_imgs=image_path,
                path_train_msks=masks_path,
                path_valid_imgs='',
                path_valid_msks='',
                train_frac=0.8,
                valid_frac=0.1,
                image_size=[512, 512],
                num_images_per_original=20,
                )
train_set, valid_set = augment_data(
    train_imgs, train_msks, valid_imgs, valid_msks, batch_size,
)

tf.data.experimental.save(train_set, train_path)
tf.data.experimental.save(valid_set, valid_path)

print('Data processed, loaded and saved.\n')
print('Training set length: ', len(train_set))
print('Validation set length: ', len(valid_set))
```
